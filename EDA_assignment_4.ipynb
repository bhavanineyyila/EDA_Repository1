#### read the Bank data

```python
bank_path=r"C:\Data Science\bank.csv"
bank_df=pd.read_csv(bank_path)
bank_df.head(10)
```

```python
bank_df=pd.read_csv(bank_path,sep=';')
bank_df
```

#### Convert cat cols list 

```python
data_types=dict(bank_df.dtypes)
cat_columns=[key for key,value in data_types.items() if value=='object']
num_columns=[key for key,value in data_types.items() if value !='object']
```

#### create a Directory value_counts_df in your python location using os.mkdirs

```python
import os 
os.getcwd()
```

```python
try:
    root_directory=os.getcwd()
    new_folder_1='EDA_4_value_counts_df'
    new_dir_1=os.path.join(root_directory,new_folder_1)
    os.makedirs(new_dir_1)
except Exception as e:
    print(e)
```

#### create frequency table Data frames in your PC

```python
for i in cat_columns:
    fre_tab=bank_df[i].value_counts()
    keys=fre_tab.keys()
    values=fre_tab.values
    cols=['Lables','Count']
    df=pd.DataFrame(zip(keys,values),
                   columns=cols)
    df.to_csv(f'{new_dir_1}\\{i}_val_con.csv',index=False)
```

#### create a Directory Graphs_matplotlib

```python
try:
    root_directory=os.getcwd()
    new_folder_2='EDA_4_Graphs_matplotlib'
    new_dir_2=os.path.join(root_directory,new_folder_2)
    os.makedirs(new_dir_2)
except Exception as e:
    print(e)
```

```python
dfs=os.listdir(new_dir_1)
dfs
```

#### Read the DataFrame and plot the matplotlib graphs and save in directory

```python
import matplotlib.pyplot as plt

for i in dfs:
    df=pd.read_csv(f"{new_dir_1}\{i}")
    plt.bar('Lables',
           'count',
           data=df)
    plt.title(f"Bar Chart {i}")
    plt.xlabel('Lable')
    plt.ylable('Count')
    plt.savefig(f"{new_dir_2}\\{i}_matplotlib.jpg")
```

```python
import seaborn as sns

for i in cat_columns:
    plt.figure(figsize=(15,6))
    order_data=bank_df[i].value_counts().keys()
    sns.countplot(data=bank_df,
                 x=i,
                 order=order_data)
    plt.title(f'{i}_Barchart')
    plt.savefig(f"{new_dir_3}\\{i}_seqaborn.jpg")
    plt.show()
```
